```python
import matplotlib.pyplot as plt
import numpy as np

# Load the image
img = plt.imread('peppers.bmp')

# Display the image
plt.imshow(img)
plt.title('Original Image')
plt.show()

# Print the shape of the image to see what it is
print("Shape of the Image: ", img.shape)

# Reshape the image into an N x 3 matrix
N = img.shape[0] * img.shape[1]
img_matrix = np.reshape(img, (N, 3))

# Print the shape of the new matrix
print("Shape of the New Matrix: ",img_matrix.shape)
```

```python
from sklearn.cluster import KMeans
import warnings

# Define the number of clusters
k = 2

# Suppress the warning and set the value of n_init
warnings.filterwarnings('ignore', message='The default value of `n_init`')
n_init = 10

# Run the k-means algorithm
kmeans = KMeans(n_clusters=k, n_init=n_init, random_state=0).fit(img_matrix)

# Get the cluster centres
cluster_centers = kmeans.cluster_centers_

# Get the labels for each pixel
labels = kmeans.labels_

# Print the cluster centres
print("Cluster Centres: ", cluster_centers)

# Print the labels
print("Labels: ", labels)
```

```python
# Replace the pixel values with the cluster center values
compressed_img = np.zeros_like(img)
for i in range(N):
    compressed_img[np.unravel_index(i, img.shape[:2])] = cluster_centers[labels[i]]

# Display the compressed image
plt.imshow(compressed_img)
plt.title('Image with ' + str(k) + ' Clusters')
plt.show()
```
